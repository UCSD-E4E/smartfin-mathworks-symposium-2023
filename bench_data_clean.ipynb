```python
import numpy as np
import pandas as pd
```

```python
bench_data = pd.read_csv("raw_data/2023.02.23_pendulum_data.csv", index_col=[0])
```

```python
bench_data
```

```python

bench_data.columns = ['time',	'xAcc',	'yAcc',	'zAcc',	'xAng',	'yAng',	'zAng',	'xMag',	'yMag',	'zMag',	'temp',	'water', 'lat',	'lon']
bench_data = bench_data.iloc[2:, :]
bench_data = bench_data.reset_index(drop=True)

bench_data.drop(["temp", "water", "lat", "lon"], inplace=True, axis=1)
```

```python
bench_data
```

```python
bench_data.drop(["xAng", "yAng", "zAng", "xMag","yMag", "zMag"], inplace=True, axis=1)
bench_data = bench_data.astype(float)
bench_data['time'] = bench_data['time'].div(1000)
bench_data['time'] = bench_data['time'] - bench_data['time'].iloc[0]
```

```python
bench_data.set_index("time", drop=True)
bench_data
bench_data.to_csv("cleaned_data/1m_positive_fast_time.csv")
```
